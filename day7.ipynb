# 数据可视化

```python
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x = 'sex',y = 'age',data = data,hue= 'sex')
plt.title("患病年龄统计")
plt.xlabel("患病年龄")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='trestbps',hue='target',data=data)
plt.title('静息血压(trestbps)')
plt.xlabel('静息血压(trestbps)')
plt.xticks(rotation =  45,ha = "right")
plt.tight_layout()
plt.show()
```

```python
data['New Trestbps'] = pd.cut(x = data['trestbps'],bins = [90,110,130,150,170,190,float('inf')],labels=['90-110','110-130','130-150','150-170','170-190','190+'])
plt.figure(figsize=(10,6))
sns.countplot(x='New Trestbps',hue='target',data=data)
plt.title('静息血压(trestbps)')
plt.xlabel('静息血压(trestbps)')
plt.xticks(rotation =  45,ha = "right")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x ='thalach',data =data)
plt.title("最高心率")
plt.xlabel("最高心率")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(x='chol',data = data,hue='sex',kde=True,element='step')
plt.title('血清胆固醇浓度(chol)')
plt.xlabel('血清胆固醇浓度(chol)')
plt.show()
```
